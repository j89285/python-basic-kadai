```python
overage = 20
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def checkadult(self):
    if self.age >= overage:
      return True
    else:
      return False

namelists = {"花子":20, "太郎":15, "父":50, "母":48}

def display_judge(name, age, result):
  print(f"{name}は{age}才なので、{result}")

for namevalue, agevalue in namelists.items():
  people = Human(namevalue, agevalue)
  if people.checkadult():
    result = "大人である"
  else:
    result = "大人ではない"

  display_judge(people.name, people.age, result)
```
